```python
# Step 1: Import Libraries
from pulp import LpMaximize, LpProblem, LpVariable, value
```

```python
# Step 2: Define the Problem
# Create a maximization problem
prob = LpProblem("Product_Profit_Optimization", LpMaximize)
```

```python
# Step 3: Define Decision Variables
# Number of units of Product A and B to produce
x_A = LpVariable('Product_A', lowBound=0, cat='Continuous')
x_B = LpVariable('Product_B', lowBound=0, cat='Continuous')
```

```python
# Step 4: Define Objective Function
# Maximize total profit
prob += 40*x_A + 50*x_B, "Total_Profit"
```

```python
# Step 5: Define Constraints
prob += 2*x_A + 4*x_B <= 100, "Labor_Constraint"
prob += 3*x_A + 2*x_B <= 90, "Material_Constraint"
```

```python
# Step 6: Solve the Problem
prob.solve()
```

```python
# Step 7: Display Results
print(f"Status: {LpStatus[prob.status]}")
print(f"Optimal units of Product A: {x_A.value()}")
print(f"Optimal units of Product B: {x_B.value()}")
print(f"Maximum Profit: ${value(prob.objective)}")
```
